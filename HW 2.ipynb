```python
import gurobipy as gp
from gurobipy import GRB
```

## Question 1:

```python
try:

    # Create a new model
    m = gp.Model("mip1")

    # Create variables
    x = m.addVar(name="x")
    y = m.addVar(name="y")
    
    # Set objective
    m.setObjective(4*x + 5*y , GRB.MAXIMIZE)

    # Add constraint: 
    m.addConstr(2*x + 3*y <= 60, "c0")
    
    # Add constraint: 
    m.addConstr(y >= x, "c1")
    
    # Add constraint: 
    m.addConstr(x >= 0, "c2")
    
    # Add constraint: 
    m.addConstr(y >= 0, "c3")

    # Optimize model
    m.optimize()

    for v in m.getVars():
        print('%s %g' % (v.varName, v.x))

    print('Obj: %g' % m.objVal)

except gp.GurobiError as e:
    print('Error code ' + str(e.errno) + ': ' + str(e))

except AttributeError:
        print('Encountered an attribute error')
```

```python
try:

    # Create a new model
    m = gp.Model("mip1")

    # Create variables
    x11 = m.addVar(name="x11")
    x12 = m.addVar(name="x12")
    x13 = m.addVar(name="x13")
    x21 = m.addVar(name="x21")
    x22 = m.addVar(name="x22")
    x23 = m.addVar(name="x23")

    # Set objective
    m.setObjective(x11 + x12 + x13 + x21 + x22 + x23, GRB.MAXIMIZE)

    # Add constraint: 
    m.addConstr(x11 + x21 >= 1.2, "c0")
    
    # Add constraint: 
    m.addConstr(x11 + x21 <= 2, "c1")
    
    # Add constraint: 
    m.addConstr(1.3*x12 + 1.2*x22 >= 1.5, "c2")
    
    # Add constraint: 
    m.addConstr(1.3*x12 + 1.2*x22 <= 2, "c3")
    
    # Add constraint: 
    m.addConstr(1.4*x13 + 1.6*x23 >= 2, "c4")
    
    # Add constraint: 
    m.addConstr(1.4*x13 + 1.6*x23 <= 3, "c5")
    
    
    # Add constraint: 
    m.addConstr(x11 + x12 + x13 <= 2, "c6")
    
    # Add constraint: 
    m.addConstr(x21 + x22 + x23 <= 3, "c7")

    # Optimize model
    m.optimize()

    for v in m.getVars():
        print('%s %g' % (v.varName, v.x))

    print('Obj: %g' % m.objVal)

except gp.GurobiError as e:
    print('Error code ' + str(e.errno) + ': ' + str(e))

except AttributeError:
    print('Encountered an attribute error')
```
